```python
#program to limit the number of cookies
num=int(input("How many cookies do you want?"))
i=1
limit=8
while i<=num:
    if i>limit:
        break
    print ("cookie",i)
    i+=1
print("You can't get more than",limit,"cookie")    
```

```python
def calculate_lcm(x=10,y=12):
    print("the value of x is",x,"the value of y is",y)
    if x>y:
        Greater = x
    else:
        Greater = y
    while(True):
        if ((Greater%x==0) and (Greater%y==0)):
            lcm=Greater
            break
        Greater+=1
    return lcm   
num1=int(input("Enter first number: "))
num2=int(input("Enter second number: "))
print("The LCM of",num1,"and" ,num2, "is",calculate_lcm(num1,num2))
```

```python
#Fibonacci series
def fib(num):
    fib_series = [0,1]
    for i in range(2,num):
        fib_series.append(fib_series[-2]+fib_series[-1])
    print(fib_series)    
fib(10)
```

```python
#fibonacci series using while  loop
n=int(input("Enter the number of terms:"))
a , b = 0, 1
count = 0
while count < n:
    print(a,end="")
    a,b = b, a+b
    count+=1
```
